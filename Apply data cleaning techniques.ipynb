**4. Performing hypothesis testing**

```python
total = df['Coal']
print (total)
```

```python
total_mean = np.mean(total)
print(total_mean)
```

**5. Performing Z Test**

```python
ztest ,pval = stests.ztest(df['Coal'], value=1000)
print(float(pval))
```

```python
if pval<0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```
